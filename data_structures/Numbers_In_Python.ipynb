```python
# iter and next methods
nums = [1, 2, 3, 4, 5]
it = iter(nums)          # returns an iterator object
print(it)

print(it.__next__())    # calling __next__() to get the first element
print(it.__next__())
print(it.__next__())
print(it.__next__())
print(it.__next__())
print(it.__next__())    # raise StopIteration error after last element
```

```python
# The Python __rlshift__() method implements the reverse bitwise left-shift operation with reflected, swapped operands.  # y << x
# (x, y) = (left_operand, right_operand)
# (x, y) = (2, 1) | 1 in binary with four bit is = 0001| When we apply "int.__rlshift__(x, y)" then it is = x.__rlshift__(y) | Hence 0001 << 2(times), 0010, 0100
result1 = int.__rlshift__(2, 1)
print(result1)
print(bin(4))
```

```python
# The Python __lshift__() method implements the reverse bitwise left-shift operation with reflected, swapped operands.  # x << y
# (x, y) = (2, 1) | 2 in binary with four bit is = 0010| When we apply "int.__lshift__(x, y)" then it is = x.__lshift__(y) | Hence 0010 << 2(times), 0100, 1000
result2 = int.__lshift__(1, 2)
print(result2)
print(bin(8))
```
